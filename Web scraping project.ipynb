## Scraping same data in a harder way

```python
from bs4 import BeautifulSoup # this module helps in web scrapping.
import requests  # this module helps us to download a web page
import pandas as pd
```

```python
#The below url contains html tables with data about world population.
url = "https://en.wikipedia.org/wiki/World_population"
```

```python
# get the contents of the webpage in text format and store in a variable called data
data  = requests.get(url).text
```

### To parse a document, pass it into the BeautifulSoup constructor, the BeautifulSoup object, which represents the document as a nested data structure:

```python
# Calling the soup or html variable will show no different outcome, but parsing in beautifulsoup will adjust for the next steps.
soup = BeautifulSoup(data, "html.parser")
```

```python
#find all html tables in the web page
tables = soup.find_all('table') # in html table is represented by the tag <table>
```

```python
# we can see how many tables were found by checking the length of the tables list
len(tables)
```

```python
import pandas as pd
```

```python
for index,table in enumerate(tables):
    if ("Current world population and latest projection according the UN. Population in (millions) and percent of the global population in that year.)" in str(table)):
        table_index = index
print(table_index) 
```

```python
#print(tables[table_index].prettify())
```

```python
population_data = pd.DataFrame(columns=["Region", "2022 (percent)", "2030 (percent)", "2050 (percent)"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        Region = col[1].text
        twenty22 = col[2].text.strip()
        Twnty30 = col[3].text.strip()
        Twnty50 = col[4].text.strip()
        population_data = population_data.append({"Region":Region,"2022 (percent)":2022, "2030 (percent)":2030, "2050 (percent)": 2050}, ignore_index=True)

population_data
```

## Scraping different data from the same website with the hard way

```python
#The below url contains html tables with data about world population.
url = "https://en.wikipedia.org/wiki/World_population"
```

```python
# get the contents of the webpage in text format and store in a variable called data
data  = requests.get(url).text
```

```python
# Calling the soup or html variable will show no different outcome, but parsing in beautifulsoup will adjust for the next steps.
soup = BeautifulSoup(data, "html.parser")
```

```python
#find all html tables in the web page
tables = soup.find_all('table') # in html table is represented by the tag <table>
```

```python
for index,table in enumerate(tables):
    if ("10 most densely populated countries)" in str(table)):
        table_index = index
print(table_index)
```

```python
population_data = pd.DataFrame(columns=["Rank", "Country", "Population", "Area", "Density"])

for row in tables[table_index].tbody.find_all("tr"):
    col = row.find_all("td")
    if (col != []):
        rank = col[0].text
        country = col[1].text
        population = col[2].text.strip()
        area = col[3].text.strip()
        density = col[4].text.strip()
        population_data = population_data.append({"Rank":rank, "Country":country, "Population":population, "Area":area, "Density":density}, ignore_index=True)

population_data
```
